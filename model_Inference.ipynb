## Importing our saved model

```python
model = tf.keras.models.load_model('emotion_detection_model.h5')
```

```python
import numpy as np

def prediction(file):
  img = tf.keras.utils.load_img(file, target_size=(48, 48, 3))
  x = tf.keras.utils.img_to_array(img)

  plt.imshow(x.astype('uint8'))

  x = np.expand_dims(x, axis=0)

  images = np.vstack([x])
  classes = model.predict(images, batch_size=10)
  idx = np.argmax(classes)
  clas = ['mawrah', 'eneg', 'takut', 'senang', 'sedih', 'terkejut', 'netral']
  print('Prediction is a {}'.format(clas[idx]))
```

```python
prediction('bayi_senang.jpeg')
```
